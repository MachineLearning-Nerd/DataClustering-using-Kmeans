```python
# Getting the values and plotting it
f1 = data['Outstate'].values
f2 = data['F.Undergrad'].values
label = data['Private'].values

# Here I have made new data with only Outside and F.Undergrad
Ndata = {'Outstate': f1, 'F.Undergrad': f2}
df = pd.DataFrame(Ndata, columns=['Outstate', 'F.Undergrad'])
print(df.head())
```

```python
# Have run the Kmeans on the df. 
kmeans = KMeans(n_clusters=2).fit(df)
centroids = kmeans.cluster_centers_

Ndata = {'Outstate': f1, 'F.Undergrad': f2, 'Private':kmeans.labels_}
dF = pd.DataFrame(Ndata, columns=['Outstate', 'F.Undergrad', 'Private'])
print(dF.head())
```

```python
# This is useing matplotlib
plt.scatter(df['Outstate'], df['F.Undergrad'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)

# This is useing matplotlib
sb.set_style('whitegrid')
sb.lmplot('Outstate','F.Undergrad',dF, hue='Private',
           palette='bright',height=6,aspect=1,fit_reg=False)
```
